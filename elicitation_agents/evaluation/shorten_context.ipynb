```python
# ---------------- Imports ----------------
import os
import json
import yaml
```

```python
# ---------------- Args ----------------
# Paths
generated_dialogues_path = "evaluation/generated-utterances-dialogue"

```

```python
# ---------------- Config ----------------
with open("../../config/config.yaml", "r") as f:
    config = yaml.safe_load(f)

proj_store = config["paths"]["proj_store"]


input_folder = os.path.join(proj_store, generated_dialogues_path)
output_folder = os.path.join(proj_store, "evaluation/generated-utterances-dialogue-short")
os.makedirs(output_folder, exist_ok=True)
```

```python
# Walk through all subdirectories and files
for root, _, files in os.walk(input_folder):
    for file in files:
        if file.endswith(".json"):
            input_path = os.path.join(root, file)
            relative_path = os.path.relpath(root, input_folder)
            output_dir = os.path.join(output_folder, relative_path)
            os.makedirs(output_dir, exist_ok=True)
            output_path = os.path.join(output_dir, file)

            # Read JSON data
            with open(input_path, "r", encoding="utf-8") as f:
                data = json.load(f)

            # Modify each dialogue
            for dialogue in data:
                dialogue["turns"] = dialogue["turns"][-3:]

            # Write shortened JSON
            with open(output_path, "w", encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False, indent=2)

```
